```python
import requests
import pandas as pd

cadastro = []
list_cadastro = []
cont = 0
tam_cpf = 11

while True:
    print('-=' * 30)
    print(F"{'CADASTRO DE PESSOA FÍSICA':^55}")
    print('-=' * 30)

    primeiro_nome = str(input('Digite seu primeiro nome: ')).upper().strip()
    while len(primeiro_nome) < 4:
        print('Nome inválido! Digite novamente!')
        primeiro_nome = str(input('Digite seu primeiro nome: ')).upper().strip()

    segundo_nome = str(input('Digite seu segundo nome: ')).upper().strip()
    while len(segundo_nome) < 4:
        print('Nome inválido! Digite novamente!')
        segundo_nome = str(input('Digite seu segundo nome: ')).upper().strip()
    
    cpf = str(input('Cpf: '))
    cpf = cpf.replace('.', '').replace('.', '').replace('-','')
    while len(cpf) != tam_cpf:
        print('CPF INVÁLIDO!')
        cpf = str(input('Cpf: '))
        cpf = cpf.replace('.', '').replace('.', '').replace('-','')

    email = str(input('Digite seu email: '))
    while not '@' in email:
        print('ERRO! Email inválido!')
        email = str(input('Digite seu email: '))

    sexo = str(input('Digite seu sexo: F/M ')).upper()[0]
    while not sexo in ['F', 'M']:
        print('Sexo inválido! Tente novamente!')
        sexo = str(input('Digite seu sexo: F/M ')).upper()[0]
        
    cep = input('Cep: ')
    cep = cep.replace('-', '').replace('.', '').replace(' ','')
    
    if len(cep) == 8:
        link = f'https://viacep.com.br/ws/{cep}/json/'
        
        requisicao = requests.get(link)
        
        dic_resquisicao = requisicao.json()
    else:
        print('CEP INVÁLIDO!')


    print(f"Cidade: {dic_resquisicao['localidade']}")
    
    print(f"UF: {dic_resquisicao['uf']}")
        
    cidade = dic_resquisicao['localidade'].upper()
    
    uf = dic_resquisicao['uf'].upper()
    
    endereco = str(input('Endereço: ')).upper()

    bairro = str(input('Bairro: ')).upper()

    cont += 1

    cadastro.append(primeiro_nome)
    cadastro.append(segundo_nome)
    cadastro.append(cpf)
    cadastro.append(email)
    cadastro.append(sexo)
    cadastro.append(cep)
    cadastro.append(endereco)
    cadastro.append(bairro)
    cadastro.append(cidade)
    cadastro.append(uf)

    list_cadastro.append(cadastro[:])
    cadastro.clear()

    resp = str(input('Deseja cadastrar mais uma pessoa? Se sim digita (S) se não digita (N): ')).upper().strip()[0]
    if resp == 'N':
        break

print()
print('-=' * 30)
print(F"{'OBRIGADA! FIM DO CADASTRO':^55}")
print('-=' * 30)
print()
print(f'Foram feitos {cont} cadastros.')
print('-=' * 30)
print()
i = 0
for i in range(len(list_cadastro)):
    print(f'Nome: {list_cadastro[i][0]}.')
    print(f'Sobre nome: {list_cadastro[i][1]}.')
    print(f'email: {list_cadastro[i][2]}.')
    print(f'Sexo: {list_cadastro[i][3]}.')
    print(f'CEP: {list_cadastro[i][4]}.')
    print(f'Endereço: {list_cadastro[i][5]}.')
    print(f'Bairro: {list_cadastro[i][6]}')
    print(f"Cidade: {list_cadastro[i][7]}")
    print(f"RJ: {list_cadastro[i][8]}")
    i += 1
    print()
    print('-=' * 30)
    print()
print(F"{'TABELA DO CADASTRO':^55}")
tabela = pd.DataFrame(list_cadastro)
display(tabela)
```
